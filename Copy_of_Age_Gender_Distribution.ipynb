```python

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=sns.load_dataset('titanic')
#histogram for continuous variable
plt.figure(figsize=(6,4))
sns.histplot(df['age'],kde=True,color='pink',bins=12)
plt.title('Age distribution from titanic dataset')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
#barchart for categorical data
plt.figure(figsize=(6,4))
sns.countplot(x='sex',data=df,palette='Set2',hue=False,legend=False)
plt.title('Gender distribution from titanic dataset')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```
